```python
def getWithSeriesName(path , label=True , figSize = (14,14) , cMap = "gray" , random = False):
    imgs_with_path = []
    
    for root , dirs , files in os.walk(path):
        for file in files:
            relPath = os.path.relpath(root , path)
            imgs_with_path.append(path + "/" + relPath + "/" + file)
    
    fig , ax = plt.subplots(nrows = 1 , ncols = 2 , figsize = figSize)
    
    label_dis = sheet_1.loc[sheet_1.ID1 == path[53:] , "classification"].values[0]
    
    for i in range(2):
        im = imageio.imread(imgs_with_path[i])
        ax[i].imshow(im , cmap = cMap)
        
        if label:
            ax[i].set_title(path[53:] + "_" + label_dis)
    
    return 
```

```python
def getRandomWithAge(age , label = True , figsize = (14,14) , cMap = "gray"):
    count_ = sheet_1[sheet_1.Age == age].shape[0]
    if(count_ != 0):
        id_list = sheet_1[sheet_1.Age == age].ID1.to_list()
        id_ = id_list[random.randint(0 , count_-1)]
        
        path = "../input/cmmd2022/TheChineseMammographyDatabase/CMMD" + "/" + id_
        
        getWithSeriesName(path , random = True , cMap = cMap)
        
    return
```

```python
def getRandomWithClass(labelClass , label = True , figsize = (14,14) , cMap = "gray"):
    count_ = sheet_1[sheet_1.classification == labelClass].shape[0]
    
    id_list = sheet_1[sheet_1.classification == labelClass].ID1.to_list()
    id_ = id_list[random.randint(0 , count_-1)]
        
    path = "../input/cmmd2022/TheChineseMammographyDatabase/CMMD" + "/" + id_
        
    getWithSeriesName(path , random = True , cMap = cMap)
        
    return
```
